```python
comorbidities = []
comorbidities_txt = []

# Open comorbidities file and read each line seperately
with open("data/comorbidities.txt", "r") as f:
    for line in f:
        line = line.strip()
        comorbidities_txt.append(line)
        if line:  # Skip empty lines
            parts = [p.strip() for p in line.split("=")]
            comorbidities.append(parts)

df_comorbidities = pd.DataFrame(comorbidities[1:])
print(df_comorbidities)
```

### Overall conclusion
The data looks pretty clean, with little null values (at least none in the CDS features). The sentiment features do have more null values. One weird thing is that one user is missing from the cohort_mapping data. To improve model performance, I will have to find a suitable solution to handling missing data, especially for neural networks.

Since the sentiment features differ drastically from tweet to tweet, I don't think mean or mode is appropriate. Imputing another static value (like -10 or 0) will likely not work either, since this will introduce more noise. Therefore, I think the best method is to use K nearest neighbours, since the sentiment features depend on the text from tweets, and the CDS features can be used to train a KNN to detect similar cases and impute a similar score.

```python
import numpy as np
import matplotlib.pyplot as plt

# Class counts per problem
counts_anxiety = [n_anxiety_0, n_anxiety_1]
counts_depression = [n_depression_0, n_depression_1]
class_labels = ['Minority Class (0)', 'Majority Class (1)']
problem_labels = ['Anxiety', 'Depression']

# Arrange the data as: rows = problems, columns = classes
data = np.array([counts_anxiety, counts_depression])

x = np.arange(len(problem_labels))  # positions for Anxiety and Depression
width = 0.35  # width of bars

fig, ax = plt.subplots()

# Bars for Class 0
bars0 = ax.bar(x - width/2, data[:, 0], width, label='Minority Class (0)', color='skyblue')
# Bars for Class 1
bars1 = ax.bar(x + width/2, data[:, 1], width, label='Majority Class (1)', color='salmon')

# Labels and title
ax.set_ylabel('Number of Samples')
ax.set_title('Class Distribution per disorder')
ax.set_xticks(x)
ax.set_xticklabels(problem_labels)
ax.legend(title='Class', loc='lower right')

# Annotate bars
for i in range(len(problem_labels)):
    ax.annotate(f'{data[i, 0]}\n({data[i, 0] / data[i].sum():.1%})',
                xy=(x[i] - width/2, data[i, 0]),
                xytext=(0, 5), textcoords="offset points",
                ha='center', va='bottom')

    ax.annotate(f'{data[i, 1]}\n({data[i, 1] / data[i].sum():.1%})',
                xy=(x[i] + width/2, data[i, 1]),
                xytext=(0, 5), textcoords="offset points",
                ha='center', va='bottom')

plt.ylim(0, max(data.flatten()) * 1.2)
plt.show()
```
